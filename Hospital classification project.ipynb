```python
# Handling the categorical data
X= pd.get_dummies(df,drop_first=True)
X
```

```python
# Splitting the dataset into training and testing
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y, test_size = 0.2, random_state = 50)
```

```python
# Apply Feature Scaling approach.
from sklearn.preprocessing import MinMaxScaler
sc= MinMaxScaler()
sc_data = sc.fit_transform(X)
sc_data
```

```python
# Model creation
# we will use regression models
# Linear Regression Model
from sklearn.linear_model import LinearRegression
slr= LinearRegression()
slr.fit(X_train,Y_train)
```

```python
y_pred= slr.predict(X_test)
y_pred
```

```python
Y_test
```

```python
# Evaluating performace of model
from sklearn.metrics import mean_squared_error,r2_score,mean_absolute_error
```

```python
mean_squared_error(Y_test, y_pred)*100
```

```python
r2_score(Y_test, y_pred)*100
# High r2_score indicates higher accuracy of the model
```

```python
np.sqrt(mean_squared_error(Y_test, y_pred))*100
```

```python
mean_absolute_error(Y_test, y_pred)*100
```

```python
# Decision Tree Regression model
# Build a Decision Tree model using criterion as Entropy
from sklearn.tree import DecisionTreeClassifier, export_text
clf = DecisionTreeClassifier(criterion='entropy')
clf.fit(X_train, Y_train)
```

```python
Y_pred2 =clf.predict(X_test)
Y_pred2
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_test,Y_pred2)
cm
```

```python
# Evaluating performace of model
from sklearn.metrics import accuracy_score, recall_score, f1_score
```

```python
accuracy_score(Y_test,Y_pred2)*100
# Higher accuracy indiactes model is highly accurate
```

```python
recall_score(Y_pred2,Y_test,average = "macro")*100
```

```python
f1_score(Y_pred2,Y_test,average = "macro")*100
```
